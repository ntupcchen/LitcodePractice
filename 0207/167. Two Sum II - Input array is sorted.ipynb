```python
class Solution:
    def twoSum(self, numbers, target: int):
        '''
        Runtime: 60 ms, faster than 88.93% of Python3 online submissions for Two Sum II - Input array is sorted.
        Memory Usage: 13.2 MB, less than 100.00% of Python3 online submissions for Two Sum II - Input array is sorted.
        '''
        l,r = 0, len(numbers)-1
        while l<r:
            print(l,r)
            if numbers[l]+numbers[r] < target:
                l += 1
            elif numbers[l]+numbers[r] > target:
                r -= 1
            else:
                return [l,r]
        return "Pair not found"
```
