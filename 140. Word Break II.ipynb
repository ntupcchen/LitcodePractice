```python
# top-down approach
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> List[str]:
        def dp(idx, memo):
            if idx in memo:
                return memo[idx]
            res = []
            for word in wordDict:
                if s[idx:].startwith(word):
                    for sentence in dp(idx+len(word),memo):
                        res.append(s[idx:idx+len(word)]+' '+sentence)
            memo[idx] = res
            return res 
                
        length_s = len(s)
        memo = {length_s:['']}
        ans = dp(0,memo)
        for i in range(len(ans)):
            ans[i] = ans[i].rstrip()
        return ans
```
